### Project: MeshMaker

#### Overview: A program that receives customers' orders for custom 3D models and animation. The customers should be able to do the following tasks for the store to function properly. As a customer, I should be able to do the following:

- Browse 3D models.
- View each model's info (image, description, price)
- Order my own custom model.
- Get an estimated price for my order.
- Cancel my order.
- Contact the MeshMaker team.
- Get a QR code for my order summary.

```python
#import needed libraries

import glob 
from IPython.display import Image, display
from PIL import Image
import qrcode
from datetime import datetime

#define a function that print the program options menu and take the user choice
def menu():

    print("|                                                                                                    |",
          "\n|  Please enter your choice number:                                                                  |",
          "\n|                                                                                                    |",
          "\n|    1- To brows our 3D models examples.                                                             |",
          "\n|    2- To order your custom model.                                                                  |",
          "\n|    3- To estimate the order price.                                                                 |",
          "\n|    4- To contact the MeshMaker team.                                                               |",
          "\n|    0- To exit the program.                                                                         |",
          "\n|                                                                                                    |")    
    
    userChoice = int(input()) #get user choice
    
    return userChoice

#define a function that will display models examples for the user
def modelsExamples():
    
    #save a list that contains models images paths
    listOfImages = [img for img in glob.glob("models\*")] #glob used here to returns an array of filenames for models images
    
    #create a dictionary that stores models info
    modelsInfo = {"name": ["Anvil", "BlueCube", "Donuts", "Glasses", "Ring", "StormyOcean"] , 
              "description": ["A 3d model of an anvil that has detailed textures and a lot of well-designed loop cuts.", "A 3d model of a -nice looking- high transparency blue cube.", "A detailed 3d model of a donut and a cup of coffee, make render of the scene and animation of sliding donut plate.", "A 3d model of a studying glasses that has custom buttons on its side, and make a render and animation of it in a simple environment that has been created with additional objects to enhance render looking.", "A 3d model of a ring that has smooth shiny looking with custom colors on its stone, and make a render of it with great lighting settings.", "A render of a full scene of a stormy ocean, including: ocean, sky, rains models"], 
              "price": [250, 200, 100, 180, 180, 80], 
              "path": listOfImages}
    

    #print all saved models info for the user
    for i in range(len(modelsInfo["name"])):
        print("Model number "+str(i+1)+": ")
        print("Name: ", modelsInfo["name"][i])
        print("Description: ", modelsInfo["description"][i])
        print("Price: ", modelsInfo["price"][i])
        #display an image for each model
        display(Image.open((modelsInfo["path"][i])).resize((200, 150)))

#define a function that will take new custom model order from the user and add it to orders dictionary for later use by meshMakers team
def orderCustomModel(ordersDict):

    while True:
        
        #get the custom model description from the user
        orderDetails = input("Please describe your model with details: ")
        print("Your 3D model order is: ", orderDetails)
        
        userOrderOption = input("Are you sure about your order? \nEnter \"Y\" to confirm your order or \"N\" to cancle: ").upper()
        
        #validate user input
        while (userOrderOption != "Y") and (userOrderOption != "N"):
                userOrderOption = input("Please enter only \"Y\" to confirm your order or \"N\" to cancle: ").upper()
                
        if userOrderOption == "Y": #means that the user want to confirm his order
            
            #save current date and time
            now = datetime.now() #datetime object containing current date and time
            #save order info in the orders dictionary for later use by meshMaker team
            ordersDict["description"].append(orderDetails)
            ordersDict["time"].append(now)
            print("\nYour order confirmed at "+str(now)+"\n\n") 
            break
            
        elif userOrderOption == "N": #means that the user want to cancle
            break
            
    return ordersDict

#define a function that will ask the user for some details about his order to give an estimate price for it
def estimatePrice():
    
    price = 0
    
    while True:
        
        print("\nPlease answer the following questions to get an estimate of your 3D model's price.\n")
        
        #ask questions and save user answer for each one
        #first question
        durationOption = int(input("* You need your model to be ready within ____? \n1- Two weeks.\n2- One month.\n3- Three months or more.\n Please enter the preferred option number: "))
        
        while (durationOption < 1 or durationOption > 3): #for validation
                durationOption = int(input("\nPlease enter only the number of your choice: "))
        #second question        
        background = input("* Do you need background to create a great looking render for your model? \nPlease enter \"Y\" for Yes or \"N\" for No: ").upper()
        
        while (background != "Y") and (background != "N"): #for validation
                background = input("\nPlease enter only \"Y\" for Yes or \"N\" for No: ").upper()
        #third question        
        animationOption = input("* Do you want to create an animation for your model? \nPlease enter \"Y\" for Yes or \"N\" for No: ").upper()
        
        while (animationOption != "Y") and (animationOption != "N"): #for validation
                animationOption = input("\nPlease enter only \"Y\" for Yes or \"N\" for No: ").upper()
        
        if animationOption == "Y":
            #fourth question
            animationDurationOption = int(input("* How long you want your animation to be? \n1- Less than one minute.\n2- One to three minutes.\n3- More than four minutes.\n Please enter the preferred option number: "))
            
            while (animationDurationOption < 1 or animationDurationOption > 3): #for validation
                    animationDurationOption = int(input("\nPlease enter only the number of your choice: "))
        
        #calculate the estimate price depending on user options
        if durationOption == 1:
            price += 50
            
        elif durationOption == 2:
            price += 30
        else:
            price += 20
            
        if background == "Y":
            price += 50
            
        if animationOption == "Y":
            price += 70
            
            if animationDurationOption == 1:
                price += 40
            elif durationOption == 2:
                price += 50
            else:
                price += 60
            
            break
            
        break #to brake from the outer loop
        
    #show the result for the calculation
    print("\nThe estimated price for your order is: ", price,"\n")
            

#define a function that will show a QR code for the support team email    
def contactTeam():
    
    print("Kindly scan the following QR code to contact the support team...")
    
    img = qrcode.make('Contact the MeshMaker team: zainabmel1421@gmail.com') # type is (qrcode.image.pil.PilImage)
    #save the generated QR code image and display it for the user
    img.save("QR_code\contact_QR_code.png")
    display(Image.open("QR_code\contact_QR_code.png").resize((200, 200)))
    
    
#define the main function that will include the main logic of the program
def main():
    
    #show welcom message
    print("\n --------------------------------  Welcome to the MeshMaker Program  -------------------------------- ",
          "\n|                                                                                                    |",
         "\n|  You can brows 3D Models, order your own custom model, and contact us at any time!                 |",
         "\n|                                                                                                    |")


    #get the name of the user
    userName = input("|What is your name?                                                                                  |"
                     +"\n\n")
    
    #use lambda that accepts one argument to greet user
    greetUser = lambda name : print('\nHey,', name,"\n")

    #call lambda function with user name
    greetUser(userName)


    userChoice = 20 #give arbitrary number to enter the loop (not 0 because 0 will exit the program later)
    ordersDict = {"description": [],"time": []} #create a dictionary that will store the users orders info later

    while userChoice!=0: #wile the user doesn't choose to exit the program

        userChoice = menu() #display the progran menu

        #if the user choose to brows our 3D models examples.
        if userChoice == 1:
            modelsExamples()

        #if the user choose to order a custom model. 
        elif userChoice == 2:
            ordersDict = orderCustomModel(ordersDict)

        #if the user choose to estimate the order price.    
        elif userChoice == 3:
            estimatePrice()

        #if the user choose to contact the MeshMaker team.   
        elif userChoice == 4:
            contactTeam()

    #call lambda function with user name
    greetUser(userName)
    print(" ----------------------- Thanks for using MeshMaker! Hope to see you again :) ----------------------- ")
    
#call the main function that will start the program
main()
```
