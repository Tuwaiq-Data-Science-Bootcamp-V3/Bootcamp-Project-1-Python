#### Overview: An online website that shows different events. The visitor should be able to do the following tasks for the store to function properly. As a visitor, I should be able to do the following:

The given code represents a cookie store program that allows users to view available cookie types, add cookies to a shopping cart, remove cookies from the cart, display the cart contents, calculate the total price of the cart, and exit the program.


The program uses a dictionary called cookies to store available cookie types and their prices. The keys represent the cookie types, and the values represent their prices. There is also a list called cart to represent the shopping cart. Each item in the cart list is a tuple containing the cookie type and the quantity.

### The code includes several functions:

- isplay_cookie_types(sort_order): This function displays the available cookie types sorted by price in either ascending or descending order, based on the sort_order parameter.

- add_to_cart(cookie_type, quantity): This function adds cookies to the cart. It checks if the given cookie_type is valid, and if so, adds the specified quantity of cookies to the cart.

- remove_from_cart(cookie_type, quantity): This function removes cookies from the cart. It searches for the given cookie_type in the cart and removes the specified quantity of cookies if they exist. It also handles cases where the quantity is insufficient or the cookie type is not found.

- display_cart(): This function displays the contents of the cart, showing the cookie types and their respective quantities.

- calculate_total_price(): This function calculates the total price of the items in the cart based on the prices specified in the cookies dictionary.

The main program loop allows users to select options from a menu by entering a number. Depending on the chosen option, the corresponding function is called to perform the desired action. The loop continues until the user selects the "Exit" option (number 7).


```python
# Cookie Store

# Dictionary to store available cookie types and their prices
cookies = {
    "Milk Chocolate Chip": 12.0,
    "Dark Chocolate Chip": 12.0,
    "Oatmeal Raisin": 11.0,
    "Crinkle Cookies": 10.5,
    "Peanut Butter": 10.0,
    "Gingerbread cookie": 10.0,
    "Butter Cookies": 10.0,
    "Snowball Cookies": 10.0,    
    "Sugar": 7.0,

}

# List to represent the shopping cart
cart = []

# Function to display the available cookie types sorted by price
def display_cookie_types(sort_order):
    sorted_cookies = sorted(cookies.items(), key=lambda x: x[1], reverse=(sort_order == "desc"))
    #sorted function is a built-in function in Python.It is used to sort iterables such as lists, tuples, or dictionaries
    
    print("\nAvailable Cookie Types:")
    for cookie_type, price in sorted_cookies:
        print(f"{cookie_type}: {price} SAR")

# Function to add cookies to the cart
def add_to_cart(cookie_type, quantity):
    if cookie_type in cookies:
        cart.append((cookie_type, quantity))
        print(f"{quantity} {cookie_type} cookie(s) added to the cart.")
    else:
        print("Invalid cookie type. Please select from the available options.")

# Function to remove cookies from the cart
def remove_from_cart(cookie_type, quantity):
    for item in cart:
        if item[0] == cookie_type:
            if item[1] >= quantity:
                item_index = cart.index(item)
                item = (item[0], item[1] - quantity)
                if item[1] == 0:
                    cart.pop(item_index)
                else:
                    cart[item_index] = item
                print(f"{quantity} {cookie_type} cookie(s) removed from the cart.")
                return
            else:
                print("Not enough quantity in the cart.")
                return
    print("Cookie type not found in the cart.")

# Function to display the contents of the cart
def display_cart():
    print("Cart Contents:")
    for item in cart:
        print(f"{item[0]} - Quantity: {item[1]}")

# Function to calculate the total price of the cart
def calculate_total_price():
    total_price = 0.0
    for item in cart:
        total_price += cookies[item[0]] * item[1]
    return total_price

# Main program loop
while True:
    print("\n=== Cookie Store ===")
    print("1. Display Available Cookie Types (sorted by price ascending)")
    print("2. Display Available Cookie Types (sorted by price descending)")
    print("3. Add Cookies to Cart")
    print("4. Remove Cookies from Cart")
    print("5. Display Cart")
    print("6. Calculate Total Price")
    print("7. Exit")

    choice = input("Enter your choice (1-7): ")

    if choice == "1":
        display_cookie_types("asc")
    elif choice == "2":
        display_cookie_types("desc")
    elif choice == "3":
        cookie_type = input("\nEnter the cookie type: ")
        quantity = int(input("Enter the quantity: "))
        add_to_cart(cookie_type, quantity)
    elif choice == "4":
        cookie_type = input("\nEnter the cookie type: ")
        quantity = int(input("Enter the quantity: "))
        remove_from_cart(cookie_type, quantity)
    elif choice == "5":
        display_cart()
    elif choice == "6":
        total_price = calculate_total_price()
        print(f"\nTotal Price: {total_price} SAR")
    elif choice == "7":
        print("\n\nThank you for shopping with us. Goodbye!")
        break
    else:
        print("\nInvalid choice. Please try again.")
```
